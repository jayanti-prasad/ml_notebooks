# Gaussian Mixture Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

def get_data (mean, cova):
   data =  np.random.multivariate_normal (mean, cova, 100)
   return data


if __name__ == "__main__":

    mean1 =  (2.0,3.0)
    cova1 = [[1.0,0.25],[0.25,1.0]]

    mean2 =  (-2.0,1.0)
    cova2 = [[1.0,-0.25],[0.25,1.0]]

    mean3 =  (1.0,-1.0)
    cova3 = [[1.0,0.5],[0.1,1.0]]

    cls1 = get_data(mean1, cova1)
    cls2 = get_data(mean2, cova2)
    cls3 = get_data(mean3, cova3)
    
    print(cls1.shape) 

    plt.scatter (cls1[:,0],cls1[:,1], label='Clsuter 2')
    plt.scatter (cls2[:,0],cls2[:,1], label='Clsuter 1')
    plt.scatter (cls3[:,0],cls3[:,1], label='Clsuter 3')
    plt.legend()
    plt.show()
    

    X = arr = np.concatenate((cls1, cls2, cls3))
    
    gm = GaussianMixture(n_components=3, random_state=0).fit(X)
    print(gm.means_)
    print(dir(gm))
    print(gm.predict([[2, 4], [-2, 1],[1,0]]))

    print(gm.weights_)
```
